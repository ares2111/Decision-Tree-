## DECISION TREE CLASSIFICATION AND VISUALIZATION


```python
# Importing libraries in Python
import pandas as pd 
import numpy as np

# Forming the iris dataframe
df=pd.read_csv('Iris.csv')
df.set_index('Id')
df.head()
```

```python
# basic info of the dataframe
df.info()
```

```python
#basic description of the various parametrs
df.describe()
```

```python
# setting feature and targets for model training
X= df[df.columns.drop(['Id','Species'])]
y= df.Species
```

```python
# Listing out the unique differrent species in the dataset
y.unique()
```

#### It is visible that there are three target options for the classifier 

```python
#importing required visualization libraries
import matplotlib.pyplot as plt
import seaborn as sns

#plotting out various parameters for better understanding of their range
plt.figure(figsize=(12,8))
plt.subplot(2,2,1)
sns.histplot(df.SepalLengthCm,kde=True,color='#008080')

plt.subplot(2,2,2)
sns.histplot(df.SepalWidthCm,kde=True,color='#008080')

plt.subplot(2,2,3)
sns.histplot(df.PetalLengthCm,kde=True,color='#008080')

plt.subplot(2,2,4)
sns.histplot(df.PetalWidthCm,kde=True,color='#008080')
plt.tight_layout()
plt.show()
```

```python
# visualizing the dataset for various parameters and classifying by species
plt.figure(figsize=(12,8))

x= df.PetalLengthCm
y= df.PetalWidthCm
plt.subplot(2,2,1)
sns.scatterplot(x,y, hue=df.Species)

x= df.SepalLengthCm
y= df.SepalWidthCm
plt.subplot(2,2,2)
sns.scatterplot(x,y, hue=df.Species)


x= df.PetalLengthCm
y= df.SepalWidthCm
plt.subplot(2,2,3)
sns.scatterplot(x,y, hue=df.Species)

x= df.SepalLengthCm
y= df.PetalWidthCm
plt.subplot(2,2,4)
sns.scatterplot(x,y, hue=df.Species)

plt.legend(loc='best')
plt.show()
```

## Setting up the model for classififcation

```python
# Importing the decision_tree_classifier from the sklearn module
from sklearn.tree import DecisionTreeClassifier
clf= DecisionTreeClassifier()
```

```python
y=df.Species
```

```python
#fitting the data for model training
clf.fit(X,y)

print('Decision Tree classifier is ready')
```

## Visualizing the Decision Tree Classifier 

```python
from sklearn import tree
from sklearn.tree import export_graphviz

fn=['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)']
cn=y

# Setting dpi = 300 to make image clearer than default
fig, axes = plt.subplots(nrows = 1,ncols = 1,figsize = (8,8), dpi=300)

tree.plot_tree(clf,
           feature_names = fn, 
           class_names=cn,
           filled = True);

# saving the figure
fig.savefig('IRIS_Decision_tree.png')
```
